```python
from datasets import load_dataset
import pandas as pd

# Load data from Hugging Face's model hub
dataset = load_dataset("avnishkr/topics_in_ai")

data = pd.concat([pd.DataFrame(dataset[split]) for split in dataset.keys()], ignore_index=True)



```

```python
from transformers import BertTokenizer, BertModel
from bert_score import BERTScorer
# Example texts
reference = "This is a reference text example."
candidate = "This is a candidate text example."
# BERTScore calculation
scorer = BERTScorer(model_type='bert-base-uncased')
P, R, F1 = scorer.score([candidate], [reference])
print(f"BERTScore Precision: {P.mean():.4f}, Recall: {R.mean():.4f}, F1: {F1.mean():.4f}")
```

```python
import pandas as pd
from bert_score import BERTScorer


# Initialize BERTScorer
scorer = BERTScorer(model_type='bert-base-uncased')

# Calculate BERTScore for each question-response pair
scores = []
for _, row in data.iterrows():
    question = row['question']
    response = row['llama3']
    P, R, F1 = scorer.score([response], [question])
    scores.append({'question': question, 'response': response, 'bert_score': F1.item()})

# Create a new DataFrame with the scores
scores_df = pd.DataFrame(scores)

# Save the DataFrame to a new CSV file
scores_df.to_csv('bert_scores.csv', index=False)
```
