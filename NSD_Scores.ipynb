```python
from sentence_transformers import SentenceTransformer
```

```python
import pandas as pd
from datasets import load_dataset
```

```python

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
# embeddings = model.encode(sentences)
```

```python
# Load data from Hugging Face's model hub
dataset = load_dataset("avnishkr/negative_samples_dataset")

data = pd.concat([pd.DataFrame(dataset[split]) for split in dataset.keys()], ignore_index=True)
```

```python
data
```

```python
# Function to split columns
def split_columns(df):
    # Split 'correct_ans' column
    correct_df = pd.DataFrame(df['correct_ans'].to_list(), columns=[f'correct_{i+1}' for i in range(10)])

    # Split 'incorrect' column
    incorrect_df = pd.DataFrame(df['incorrect_ans'].to_list(), columns=[f'incorrect_{i+1}' for i in range(10)])

    # Concatenate the new columns with original DataFrame
    df = pd.concat([df['question'], df["llama3"], correct_df, incorrect_df], axis=1)
    return df
```

```python
# Split columns
df = split_columns(data)
```

```python

# Display the DataFrame
df
```

```python
def create_embeddings(df):
    embeddings = []
    columns = df.columns[1:]  # Exclude the question column
    for col in columns:
        # Get embeddings for each entry in the column
        col_embeddings = []
        for entry in df[col].astype(str):
            embedding = model.encode(entry)
            col_embeddings.append(embedding)
        embeddings.append(col_embeddings)
    return embeddings
```

```python
import numpy as np
```

```python
# Create embeddings for each column except the question column
embeddings = create_embeddings(df)
```

```python
# Flatten the embeddings to have a shape of (n_entries, 384)
for i, col in enumerate(embeddings):
    embeddings[i] = np.array(col).reshape(-1, 384)

# Create a new DataFrame with embeddings
new_df = pd.DataFrame()
new_df["question"] = df["question"]
for i, col in enumerate(embeddings):
    new_df[df.columns[i+1]] = col.tolist()
```

```python
# Save DataFrame to CSV
new_df.to_csv("embeddings.csv", index=False)
```

```python
def calculate_similarity(llm_ans, answers):
    # Get embeddings for the question and answers
    # llm_ans_embedding = model.encode(llm_ans)
    # answer_embeddings = model.encode(answers)

    # Calculate cosine similarity between question and answers
    # print(np.array(answers).shape)
    # print(np.array(llm_ans).shape)
    similarities = cosine_similarity([llm_ans], [answers])[0]

    # Return maximum similarity
    return max(similarities)
```

```python
# Calculate lambda value
def calculate_lambda(row):
    llm = row['llama3']
    correct_ans = [row[f'correct_{i+1}'] for i in range(10)]
    incorrect_ans = [row[f'incorrect_{i+1}'] for i in range(10)]

    max_correct = max(calculate_similarity(llm, ans) for ans in correct_ans)
    max_incorrect = max(calculate_similarity(llm, ans) for ans in incorrect_ans)

    return max_correct - max_incorrect
```

```python
# Create a new DataFrame for lambda values
lambda_df = pd.DataFrame()
lambda_df['question'] = new_df['question']
lambda_df['lambda'] = new_df.apply(calculate_lambda, axis=1)

# Save to CSV
lambda_df.to_csv("lambda_values.csv", index=False)
```
